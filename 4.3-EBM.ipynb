# Explainable Boostgin Classifier

We are going to use the InterpretML library provided by Microsoft https://github.com/interpretml/interpret

```python

from interpret.glassbox import ExplainableBoostingClassifier
from interpret import show
import pandas as pd

from interpret import set_visualize_provider
from interpret.provider import InlineProvider
set_visualize_provider(InlineProvider())
```

```python
from sklearn.model_selection import train_test_split

df = pd.read_csv("./dataset/users_df_dataset_cleaned_with_indicators.csv")
columns = [
#    "user_id"  ,
#    "name"  ,
#    "lang"  ,
    "bot"  ,
#    "created_at" ,
    "statuses_count" ,
    "account_age_in_days" ,
    "number_of_tweets" ,
    "account_average_tweets_per_day" ,
    "avg_tweets_per_actual_day" ,
#    "day_with_most_tweets" ,
    "max_number_of_tweets_in_a_day",
    "entropy_for_day",
    "entropy_for_hour",
    "entropy_for_minute",
    "avg_hashtags",
    "avg_text_length",
    "avg_mentions",
    "avg_special_char_in_text",
    "total_likes",
    "avg_favorite_count",
    "total_replies",
    "avg_reply_count",
    "total_retweet_count",
    "account_discussion_creation_ratio",
    "tweet_num_likes_ratio",
    "tweet_num_replies_ratio",
    "entropy_original_text",
    "entropy_text",
    "mean_inactive_period_length_in_seconds",
    "median_inactive_period_length_in_seconds",
    "mode_inactive_period_length_in_seconds",
    "mode_count"
]

df = df[columns]

labels = df.pop("bot")

X_train, X_test, y_train, y_test = train_test_split(df.values, labels.values, test_size=0.20, random_state=1)
```

```python
ebm = ExplainableBoostingClassifier(random_state=1)
ebm.fit(X_train, y_train)

ebm_global = ebm.explain_global()
show(ebm_global)

#ebm_local = ebm.explain_local(X_test, y_test)
#show(ebm_local)
```
